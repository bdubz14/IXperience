## Problem 1

*The file `crime_data.txt` contrains data on types of crimes commited in the US from 1960 - 2016. Complete the following problems using this data.*

Load the data into a pandas DataFrame, using the column names in the file and the column titled “Year” as the index. Make sure to skip lines that don’t contain data.

```python
crimes = pd.read_csv('crime_data.txt', sep=',', header=1)
crimes = crimes.set_index('Year', drop=False)
crimes.head()
```

Insert a new column into the data frame that contains the crime rate by year (the ratio of “Total” column to the “Population” column).

```python
crimes['Rate'] = crimes.Total / crimes.Population
crimes.head()
```

List the 5 years with the highest crime rate in descending order.

```python
crimes = crimes.set_index('Rate', drop=False)
crimes = crimes.sort_index(ascending=False)
crimes.head(5)
```

Calculate the average number of total crimes as well as burglary crimes between 1960 and
2012.

```python
total = sum(crimes['Total'])
avg_total = total / 53
burglary = sum(crimes['Burglary'])
avg_burglary = burglary / 53
print('Total average: ' + str(int(avg_total)) + ', Burglary average: ' + str(int(avg_burglary)))
```

Find the years for which the total number of crimes was below average, but the number
of burglaries was above average.

```python
crimes[(crimes.Total < avg_total) & (crimes.Burglary > avg_burglary)].Year
```

Select the Population, Violent, and Robbery columns for all years in the 1980s, and save this smaller data frame to a CSV file crime_subset.csv

```python
eighties = crimes[(crimes.Year >= 1980) & (crimes.Year <= 1989)]
eighties[['Year', 'Population', 'Violent', 'Robbery']].to_csv('crime_subset.csv', index=False)
```

## Problem 2

*The file `titanic.csv` has various features on who survived the accident.  Complete the following problems using this dataset.

Read the data into a DataFrame. Use the first row of the file as the column labels, but do not use any of the columns as the index.

```python
titanic = pd.read_csv('titanic.csv')
titanic
```

Drop the columns "Sibsp", "Parch", "Cabin", "Boat", "Body", and "home.dest"

```python
titanic.drop(['Sibsp', 'Parch', 'Cabin', 'Boat', 'Body', 'home.dest'], inplace=True, axis=1)
titanic
```

Replace null entries in the "Age" column with the average age.

```python
avg_age = sum(titanic[titanic.Age.notnull()]['Age']) / 1309
titanic.Age.fillna(avg_age)
titanic
```

Save the new DataFrame as titanic_clean.csv

```python
titanic.to_csv('titanic_clean.csv', index=False)
```

How many people survived? What percentage of passengers survived?

```python
survived = 0
for person in titanic['Survived']:
    if person == 1.0:
        survived += person
percent = survived / 1309 * 100
print(str(percent) + '% of passengers survived totaling to ' + str(survived) + ' survivors.')
```

What was the average price of a ticket? How much did the most expensive ticket cost?


```python
fare_sum = 0
for fare in titanic['Fare']:
    if not pd.isna(fare):
        fare_sum += fare
avg_fare = fare_sum / 1309
max_fare = titanic['Fare'].max()
print('Average price: ' + str(avg_fare) + ', Max price: ' + str(max_fare))
```

How old was the oldest survivor? How young was the youngest survivor? What about
non-survivors?

```python
oldest_surv = titanic[titanic.Survived == 1.0]['Age'].max()
youngest_surv = titanic[titanic.Survived == 1.0]['Age'].min()
old_non_surv = titanic[titanic.Survived == 0.0]['Age'].max()
young_non_surv = titanic[titanic.Survived == 0.0]['Age'].min()
print('Oldest survivor: {}\nYoungest survivor: {}\nOldest non-survivor: {}\nYoungest non-surivor: {}'.format(
    str(oldest_surv), str(youngest_surv), str(old_non_surv), str(young_non_surv)))
```
