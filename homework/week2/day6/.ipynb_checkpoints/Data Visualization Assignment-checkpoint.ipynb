# Data Visualization assignment

This one is simple, you have to plot a map of population by country for the year 2016.

For this you have to find:

- Population data 
- A country level geojson file.

You can use either one of the map plotting libraries we used in class.

```python
geojson = 'data/world.geo.json/countries.geo.json'
data = 'data/API_SP.POP.TOTL_DS2_en_csv_v2_9984924/API_SP.POP.TOTL_DS2_en_csv_v2_9984924.csv'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv(data)
df = df.fillna(0)
df['2016'] = df['2016'].astype(int)
df.head()
```

```python
clean_df = df[['Country Name', '2016']]
clean_df.head()
```

```python
import json
with open(geojson, errors="ignore") as fname:
    geo_data = json.load(fname)
geo_data['features'][0]
```

```python
import folium
m = folium.Map(location=[38.537327, -97.672899], zoom_start=2,
              tiles='Cartodb Positron')

m.choropleth(
    geo_data=geo_data,
    name='country_pop',
    data=clean_df,
    columns=["Country Name", "2016"],
    key_on="feature.properties.name",
    fill_color="YlOrRd",
    fill_opacity=0.9,
    line_opacity=0.2,
    legend_name="2016 World Population by Country",
)


folium.LayerControl().add_to(m)
m
```
