```python
import pandas as pd
```

```python
baseball = pd.read_csv('example_data/baseball_clean.csv')
```

```python
baseball.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
baseball_model = LinearRegression()
```

```python
names_indep_var = ['SO','FBv','SLv','CBv','CHv','HR/FB','Soft%','Med%','Hard%']
```

```python
baseball_model.fit(X=baseball[names_indep_var], y=baseball["BA"])
```

```python
baseball["BA_pred"] = baseball_model.predict(baseball[names_indep_var])
```

```python
baseball
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
plt.rcParams["figure.figsize"] = (6, 6)
```

```python
plt.plot(baseball.Season, baseball.BA, 'o')
plt.plot(baseball.Season, baseball.BA_pred)
plt.show()
```
