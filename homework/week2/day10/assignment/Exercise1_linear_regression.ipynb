### Exercise 1. Linear Regression

This is a very simple exercise, we are going to load the movies dataset, and predict the variable `revenue` using the numerical variables.

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("../data/movies.1.initial_process.csv")
```

```python
movies.head()
```

### Select the numerical columns and replace their missing values with 0

```python
numerical_data = movies.select_dtypes(['int','float']).fillna(0).astype(np.float32)
numerical_data.head()
```

### Create a linear regression model and train it on the numerical variables, using the variable `revenue` as the target variable.

```python
from sklearn.linear_model import LinearRegression
```

```python
movie_model = LinearRegression()
```

```python
names_independent_var = ['budget','popularity','runtime','vote_average','vote_count']
```

```python
movie_model.fit(X=numerical_data[names_independent_var], y=numerical_data["revenue"])
```

### Create revenue predictions and assign them to a new column on the movies dataframe named `revenue_pred`.

```python
movies["revenue_pred"] = movie_model.predict(numerical_data[names_independent_var])
```

```python
movies.head()
```

### Print the revenue and revenue_pred columns for the first 10 rows

```python
movies[['revenue','revenue_pred']].head(10)
```

```python
normalized_model  = LinearRegression(normalize=True)
normalized_model.fit(X=numerical_data[names_independent_var],
                       y=numerical_data["revenue"])
movies["revenue_pred2"] = normalized_model.predict(numerical_data[names_independent_var])
```
